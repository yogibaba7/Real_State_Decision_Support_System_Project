```python
# Price column
l1 = l1[l1['price']!='Price on Request']
l1['price'].value_counts()
```

```python
l1 = l1[l1['price']!="₹7.2 - 45 Lac"]
```

```python

def convert_to_lakh(value):
    if pd.isna(value):
        return None

    # Remove rupee symbol and spaces
    value = value.replace("₹", "").strip()


    # Handle Crores
    if "Cr" in value:
        num = float(value.replace("Cr", "").strip())
        return num * 100  # convert Cr to Lakh

    # Handle Lakhs
    if "Lac" in value or "Lakh" in value:
        num = float(value.replace("Lac", "").replace("Lakh", "").strip())
        return num


# Apply conversion
l1['price'] = l1["price"].apply(convert_to_lakh)


```

```python
# price_per_sqrt
l1['price_per_sqft'].value_counts()
```

```python
def clean_price_sqft(value):
    try:
        # remove ₹, commas, /sqft, spaces
        value = (
            value.replace("₹", "")
                 .replace("/sqft", "")
                 .replace(",", "")
                 .strip()
        )
        return float(value)
    except:
        return None

# Apply cleaning
l1["price_per_sqft"] = l1["price_per_sqft"].apply(clean_price_sqft)
```

```python
n = l1[l1['price_per_sqft']==0.0]
n
```

```python
l1.drop(n.index,inplace=True)
```

```python
# total_area
l1['total_area'] = l1['total_area'].str.replace("sqft"," ").str.replace(",","").str.strip().astype('int')
```

```python
# area_sqm
l1['area_sqm'] = l1['area_sqm'].str.replace("(","").str.replace("sqm","").str.replace(")","").str.replace(",","").str.strip().astype('int')
```

```python

# Split values and create dummy columns
df_overlook = l1["Overlooking"].str.get_dummies(sep=",")

# Select only the required 5 columns (ensure consistent order)
cols = ["Club", "Main Road", "Others", "Park/Garden", "Pool"]
for col in cols:
    if col not in df_overlook.columns:
        df_overlook[col] = 0  # add missing columns

# Convert 0/1 → Yes/No
df_overlook = df_overlook[cols].replace({1: "Yes", 0: "No"})


# Merge back into original dataframe
l1 = pd.concat([l1, df_overlook], axis=1)
```
