```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

cols = [
    'study room',
    'servant room',
    'store room',
    'pooja room',
    'others',
    'view floor plan'
]

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, col in enumerate(cols):
    
    data = (
        df.groupby('property_type')[col]
        .mean()
        .to_frame()
        .T
    )

    sns.heatmap(
        data,
        annot=True,
        cmap='YlOrRd',
        cbar=True,
        ax=axes[i]
    )
    
    axes[i].set_title(f'{col} vs PropertyType')
    axes[i].set_xlabel('Property Type')
    axes[i].set_ylabel('')

plt.suptitle('PropertyType vs Amenities Heatmaps', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# addition rooms are most offered in houses then flats
# view floor plan provided only in flats
```
