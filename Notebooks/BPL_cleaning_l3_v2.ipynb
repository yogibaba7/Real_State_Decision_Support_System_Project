```python
def Area(data):
    if not pd.isnull(data['built_up_area']):
        return data['built_up_area']
    elif not pd.isnull(data['carpet_area']):
        return data['carpet_area']
    else:
        return data['effective_area_sqft']
```

```python
area = df[['built_up_area','carpet_area','effective_area_sqft']].apply(Area,axis=1)
df.insert(11,'mainarea',area)
```

```python
df.head()
```

```python
price = (df['mainarea']  * df['price_per_sqft']) / 100000
df.insert(4,'price',price)
```
