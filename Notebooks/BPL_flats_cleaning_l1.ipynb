```python
import pandas as pd 
import numpy as np 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
t1 = pd.read_csv("bhopal_flats1.csv")
t2 = pd.read_csv("bhopal_flats2.csv")
t3 = pd.read_csv("bhopal_flats3.csv")
t4 = pd.read_csv("bhopal_flats4.csv")
t5 = pd.read_csv("bhopal_flats5.csv")
t6 = pd.read_csv("bhopal_flats6.csv")
t8 = pd.read_csv("bhopal_flats8.csv")
t9 = pd.read_csv("bhopal_flats9.csv")
t10 = pd.read_csv("bhopal_flats10.csv")
t11 = pd.read_csv("bhopal_flats11.csv")
```

```python
t = pd.concat([t1,t2,t3,t4,t5,t6,t8,t9,t10,t11])
print('shape',t1.shape[0]+t2.shape[0]+t3.shape[0]+t4.shape[0]+t5.shape[0]+t6.shape[0]+t8.shape[0]+t9.shape[0]+t10.shape[0]+t11.shape[0])
print("t",t.shape)
```

```python
# check for duplicates
t.duplicated().sum()
```

```python
# drop duplicates
t = t.drop_duplicates()
```

```python
t.shape
```

```python
t['furnishing'] = t['furnishing'].replace({"[]":np.nan})
```

```python
t['features'] = t['features'].replace({"[]":np.nan})
```

```python
# check for missing values
t.isnull().sum()
```

```python
t[t.isnull().sum(axis=1)>10].shape
```

```python
t = t[t.isnull().sum(axis=1)<10]
```

```python
t.shape
```

```python
t.head()
```

# Society

```python
t['society'].value_counts()
```

```python
t['society'].value_counts().shape
```

```python
t['society'] = t['society'].str.strip().str.lower().str.replace(",","").str.replace('bhopal','').str.strip()
```

```python
t['society'].value_counts().shape
```

```python
t = t[t['price_outer']!='Price on Request']
```

```python
t['price_outer'].value_counts()
```

```python
def price_treatment(text):
    t = text.replace('₹',"").strip()
    if "Cr" in t:
        return float(t.replace("Cr","").strip())*100
    else:
        return float(t.replace("Lac","").strip())
        
        
t['price_outer'] = t['price_outer'].apply(price_treatment)
```

```python
t.head()
```

# price_per_sqft

```python
t['price_per_sqft'].value_counts()
```

```python
t["price_per_sqft"] = t["price_per_sqft"].str.replace("₹","").str.replace(",","").str.replace("/sqft","").astype('int')
```

```python
t.head()
```

# Possesion 

```python
t['possesion'].value_counts()
```

# Configuration

```python
t['Configuration'].value_counts()
```

```python
t['Configuration'] = t['Configuration'].str.replace(
    r'\nView Floor Plan',
    ' with View Floor Plan',
    regex=True
)


config_split = t['Configuration'].str.split('with',n=1,expand=True)
rooms = config_split[0].str.split(",",expand=True)

```

```python
t[['bedrooms','bathrooms','balconies']] = rooms.rename(columns={0:"bedrooms",1:"bathrooms",2:"balconies"})
```

```python
t.head()
```

```python
t['additional_room'] = config_split[1]
```

```python
t.head()
```

# Bedrooms

```python
t['bedrooms'].value_counts()
```

```python
t['bedrooms'] = t['bedrooms'].str.replace('Bedrooms',"").str.replace('Bedroom',"").astype('int')
```

```python
t['bedrooms'].value_counts()
```

# Bathrooms

```python
t['bathrooms'].value_counts()
```

```python
t['bathrooms'] = t['bathrooms'].str.replace('Bathrooms',"").str.replace("Bathroom","").astype('int')
```

```python
t['bathrooms'].value_counts()
```

# Balconies

```python
t['balconies'].value_counts()
```

```python
t['balconies'] =  t['balconies'].str.replace('Balcony',"").str.replace('Balconies',"").str.strip().str.replace('\nView Floor Plan',"").str.replace("No","0").str.replace("3+","3").astype('int')
```

```python
t['balconies'].value_counts()
```

# additional rooms

```python
t['additional_room'].value_counts()
```

```python
t['additional_room'] = t['additional_room'].str.lower().fillna('Not Available')
```

```python
t.head()
```

```python
t['Floor Number'].value_counts()
```

```python
t['Floor_On'] = t['Floor Number'].str.extract(r'(\d+)(?:st|nd|rd|th)', expand=False).fillna(0).astype('int')
```

```python
t['Total_Floors'] = (
    t['Floor Number']
    .str.extract(r'of\s+(\d+)', expand=False)
    .astype(int)
)
```

```python
t['Total_Floors'].value_counts()
```

# Facing

```python
t['Facing'].value_counts()
```

```python
t['Facing'].isnull().sum()
```

```python
t['Facing'] = t['Facing'].fillna('NA')
```

```python
t.head()
```

```python
t['area'] = round(t['price_outer']*100000/t['price_per_sqft'],2)
```

```python
t.head(2)
```

```python
t = t.drop(columns=['Configuration','Possession in','Price','Address','ratings'])
```
