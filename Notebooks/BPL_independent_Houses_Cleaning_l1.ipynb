```python
t['furnishing'] = t['furnishing'].replace({"[]":np.nan})
```

# Society

```python
t['society'].value_counts()
```

```python
t['society'].value_counts().shape
```

```python
t['society'] = t['society'].str.strip().str.lower().str.replace(",","").str.replace('bhopal','').str.strip()
```

```python
t['society'].value_counts().shape
```

```python
t = t[t['price_outer']!='Price on Request']
```

```python
t['price_outer'].value_counts()
```

```python
def price_treatment(text):
    t = text.replace('₹',"").strip()
    if "Cr" in t:
        return float(t.replace("Cr","").strip())*100
    else:
        return float(t.replace("Lac","").strip())
        
        
t['price_outer'] = t['price_outer'].apply(price_treatment)
```

```python
t.head()
```

# price_per_sqft

```python
t['price_per_sqft'].value_counts()
```

```python
t["price_per_sqft"] = t["price_per_sqft"].str.replace("₹","").str.replace(",","").str.replace("/sqft","").astype('int')
```

```python
t.head()
```

# Possesion

```python
t['possesion'].value_counts()
```

# Configuration

```python
t['Configuration'].value_counts()
```

```python
config_split = t['Configuration'].str.split('with',n=1,expand=True)
rooms = config_split[0].str.split(",",expand=True)

```

```python
t[['bedrooms','bathrooms','balconies']] = rooms.rename(columns={0:"bedrooms",1:"bathrooms",2:"balconies"})
```

```python
t.head()
```

```python
t['additional_room'] = config_split[1]
```

```python
t.head()
```

# bedrooms

```python
t = t[~t['bedrooms'].str.contains('Bathroom')]
```

```python
t['bedrooms'].value_counts()
```

```python
t['bedrooms'] = t['bedrooms'].str.replace('Bedrooms',"").str.replace('Bedroom',"").astype('int')
```

```python
t.head()
```

# bathrooms

```python
t['bathrooms'].value_counts()
```

```python
t['bathrooms'] = t['bathrooms'].str.replace('Bathrooms',"").str.replace("Bathroom","").astype('int')
```

# balconies

```python
t['balconies'].value_counts()
```

```python
t['balconies'] = t['balconies'].str.replace('Balcony',"").str.replace('Balconies',"").str.strip().str.replace('\nView Floor Plan',"").str.replace("No","0").str.replace("3+","3").astype('int')
```

```python
t.head()
```

# additional rooms

```python
t['additional_room'].value_counts()
```

```python
t['additional_room'] = t['additional_room'].str.replace('\nView Floor Plan',"").str.lower().fillna('Not Available')
```

```python
t.head()
```

```python
t['Total Floors'].isnull().sum()
```

```python
t['Total Floors'] = t['Total Floors'].str.replace('Floors',"").str.strip()
```
