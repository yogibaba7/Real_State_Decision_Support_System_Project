# 4. price_outer

```python
df['price_outer'].isnull().sum()
```

```python
df['price_outer'].describe()
```

```python
# median = 55 lakh and mean = 91 lakh 
# mean influenced by outliers
# variability in price high standard deviation 1.2 crore
# 75% data lies within 1 crore.
```

```python
sns.histplot(df['price_outer'], kde=True, bins=50)
```

```python
# right skew distribution
```

```python
sns.boxplot(x=df['price_outer'], color='lightgreen')
plt.grid()
```

```python
# potenial outliers
```

```python
# Skewness and Kurtosis
skewness = df['price_outer'].skew()
kurtosis = df['price_outer'].kurt()

print(skewness,kurtosis)
```

```python
# highly right skew data and heavy tails means extreme outliers.
```

```python
# Quantile Analysis
quantiles = df['price_outer'].quantile([0.01, 0.05, 0.95, 0.99])

quantiles
```

```python
# 1% property price below 10 lakh
# 5% property price below 16 lakh
# 95% property price below 2.9 crore
# 99% property price below 6 crore
```

```python
# Identify potential outliers using IQR method
Q1 = df['price_outer'].describe()['25%']
Q3 = df['price_outer'].describe()['75%']
IQR = Q3 - Q1

IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(lower_bound, upper_bound)
```

```python
outliers = df[(df['price_outer'] < lower_bound) | (df['price_outer'] > upper_bound)]
outliers.shape
```

```python
outliers['price_outer'].describe()
```

```python
quantiles = outliers['price_outer'].quantile([0.01, 0.05, 0.95, 0.99])

quantiles
```

```python
# in outlier there are some outlier to
```

```python
# 152 potential outlier
# outliers median 3 crore mean 3.7 crore 
# outlier range 2 crore to 17 crore
```

```python
# price binning
bins = [0, 50, 100, 200, 300, 500, 1000, 2000]
bin_labels = ["0-50", "50-100", "100-200", "200-300", "300-500", "500-1000", "1000-2000"]
pd.cut(df['price_outer'], bins=bins, labels=bin_labels, right=False).value_counts().sort_index().plot(kind='bar')
```

```python
# majority property price fall within 2 crore.
```

```python
# ecdf plot
ecdf = df['price_outer'].value_counts(normalize=True).sort_index().cumsum()
plt.plot(ecdf.index, ecdf, marker='.', linestyle='none')
plt.grid()
```

```python
# ecdf shows 90% property prices falls within 2.5 crore
```

```python
plt.figure(figsize=(15, 6))

# Distribution plot without log transformation
plt.subplot(1, 2, 1)
sns.histplot(df['price_outer'], kde=True, bins=50, color='skyblue')
plt.title('Distribution of Prices (Original)')
plt.xlabel('Price (in Crores)')
plt.ylabel('Frequency')

# Distribution plot with log transformation
plt.subplot(1, 2, 2)
sns.histplot(np.log1p(df['price_outer']), kde=True, bins=50, color='lightgreen')
plt.title('Distribution of Prices (Log Transformed)')
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# log transform turning price skewed distribution into approxiametlly in normal. 
```

```python
skewness = np.log1p(df['price_outer']).skew()
kurtosis = np.log1p(df['price_outer']).kurt()

print(skewness,kurtosis)
```

```python
plt.figure(figsize=(15, 6))

# Distribution plot without log transformation
plt.subplot(1, 2, 1)
sns.boxplot(df['price_outer'], color='skyblue')
plt.title('Distribution of Prices (Original)')
plt.xlabel('Price (in Crores)')
plt.ylabel('Frequency')

# Distribution plot with log transformation
plt.subplot(1, 2, 2)
sns.boxplot(np.log1p(df['price_outer']), color='lightgreen')
plt.title('Distribution of Prices (Log Transformed)')
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 12))

# Create a subplot of pie charts for each room type
for idx, room in enumerate(['study room','servant room','store room','pooja room','others','view floor plan'], 1):
    ax = plt.subplot(2, 3, idx)
    df[room].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, ax=ax)
    plt.title(f'Distribution of {room.title()}')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```
